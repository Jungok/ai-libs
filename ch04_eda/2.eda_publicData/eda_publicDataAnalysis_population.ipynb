```python
import csv
f =open('age.csv')
data = csv.reader(f)
next(data)
home = []  #입력 받은 지역의 데이터를 저장할 리스트 생성
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
for row in data :
    if name in row[0] : #입력 받은 지역의 이름이 포함된 행 찾기
        for i in row[3:] : #3번 인덱스 값부터 슬라이싱
            home.append(int(i)) #입력 받은 지역의 데이터를 home에 저장
print(home) #home에 저장된 데이터 출력
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')  # 격자 무늬 스타일 지정 
plt.figure(figsize = (10,5), dpi=100)    
plt.rc('font', family ='Malgun Gothic')
plt.title(name +' 지역의 인구 구조')

plt.plot(home) 
plt.show()
```

## 전국 인구 현황 공공데이터 분석 : 남녀 성별 인구 분포 분석

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
m = []
f = []
for row in data :
    if name in row[0] :
        for i in range(0,101) :
#            m.append(int(row[i+3]))   
            m.append(-int(row[i+3]))
            f.append(int(row[-(i+1)]))
f.reverse()
```

```python
print('male:', m)
print('female:', f)
```

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
m = []
f = []
for row in data :
    if name in row[0] :
        for i in range(0,101) :  
            m.append(-int(row[i+3]))
            f.append(int(row[-(i+1)]))
f.reverse()
print('male:', m[:20])
print('female:', f[:20])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize = (5,5), dpi=100) 
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize = (5,5), dpi=100) 
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + ' 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()
```

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')
m = []
f = []
for row in data :
    if name in row[0] :
        for i in row[3:104] : 
            m.append(-int(i)) # 마이너스 부호를 넣어서 음수로 변경
        for i in row[106:] :            
            f.append(int(i))
```

```python
print('male:', m)
print('female:', f)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize = (5,5), dpi=100) 
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + ' 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()
```
